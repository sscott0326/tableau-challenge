```python
import pandas as pd
```

```python
data_2019_01 = pd.read_csv("Resources/raw_data/201901-citibike-tripdata.csv")
data_2019_02 = pd.read_csv("Resources/raw_data/201902-citibike-tripdata.csv")
data_2019_03 = pd.read_csv("Resources/raw_data/201903-citibike-tripdata.csv")
data_2019_04 = pd.read_csv("Resources/raw_data/201904-citibike-tripdata.csv")
data_2019_05 = pd.read_csv("Resources/raw_data/201905-citibike-tripdata.csv")
```

```python
data_2019_01["Month"] = "January"
data_2019_02["Month"] = "February"
data_2019_03["Month"] = "March"
data_2019_04["Month"] = "April"
data_2019_05["Month"] = "May"
```

```python
files_2019 = [data_2019_01,
              data_2019_02,
              data_2019_03,
              data_2019_04,
              data_2019_05]
```

```python
for file in files_2019:
    file.drop(['starttime',
               'stoptime',
               'start station id',
               'end station id'], axis=1,inplace=True)
    file.drop(file[file['birth year'] < 1939].index, inplace=True)
    file.drop(file[file['gender'] == 0].index, inplace=True)
    file['gender'].replace({1: "Male", 2:"Female"}, inplace=True)
```

```python
data_2019_01.to_csv("Resources/clean_data/data_2019_01.csv")
data_2019_02.to_csv("Resources/clean_data/data_2019_02.csv")
data_2019_03.to_csv("Resources/clean_data/data_2019_03.csv")
data_2019_04.to_csv("Resources/clean_data/data_2019_04.csv")
data_2019_05.to_csv("Resources/clean_data/data_2019_05.csv")
```
